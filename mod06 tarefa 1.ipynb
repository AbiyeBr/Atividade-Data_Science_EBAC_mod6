####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.close('all')

fig , ax = plt.subplots(1 , 2 , layout = 'constrained')

sns.countplot(x = 'APGAR1' , data = df_sinasc , ax = ax[0])
sns.countplot(x = 'APGAR5' , data = df_sinasc , ax = ax[1])

plt.show()
```

```python
fig , ax = plt.subplots()

sns.countplot(x = 'SEXO' , data = df_sinasc)

plt.show()
```

```python
plt.close('all')
fig , ax = plt.subplots()

sns.countplot(x = 'GESTACAO' , data = df_sinasc)
ax.set_xticklabels(ax.get_xticklabels(), rotation=10, ha='right')


plt.show()
```

```python
plt.close('all')
fig , axes = plt.subplots()

sns.countplot(x = 'ESTCIVMAE' , data = df_sinasc)

plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
fig , axes = plt.subplots()

sns.barplot(x = 'CONSULTAS' , y = 'APGAR5' , data = df_sinasc)
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig , ax = plt.subplots(2 , 2 , layout = 'constrained')

sns.countplot(x = 'SEXO' , data = df_sinasc , ax = ax[0 , 0])
sns.countplot(x = 'GESTACAO' , data = df_sinasc , ax = ax[0 , 1])
sns.countplot(x = 'ESTCIVMAE' , data = df_sinasc , ax = ax[1 , 0])
sns.countplot(x = 'CONSULTAS' , data = df_sinasc , ax = ax[1 , 1])

ax[0 , 1].set_xticklabels(ax[0 , 1].get_xticklabels(), rotation=25, ha='right')

plt.show()
```
